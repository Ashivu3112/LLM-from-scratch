# **Simplified Attention**

```python
import torch
import numpy as np
```

```python
inputs = torch.tensor(
[[ 0.32,-0.88 , 1.25 , 0.04 ,-0.52],
 [ 0.97, 0.10 ,-0.33 , 0.45 , 0.21],
 [-0.41, 1.17 , -0.73 , -0.62 , 0.87],
 [ 0.14, -0.55,  0.07 , 0.98 , 0.65],
 [-0.89, 0.34 , 0.41 , -0.21 , 0.03]]
)
```

```python
attention_scores = torch.mm(inputs, inputs.T)
attention_scores
```

```python
attention_weights = torch.softmax(attention_scores, dim=-1) # These are also probabilities how much a query attends to other tokens
print(attention_weights)
print(attention_weights[0].sum())
```

```python
context_vector = torch.mm(attention_weights, inputs) # enriched embedding vector having information of semantics.
context_vector
```

```python
context_vector2 = attention_weights @ inputs # enriched embedding vector having information of semantics.
context_vector2
```
